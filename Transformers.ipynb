```python
from transformers import pipeline

# Load pipelines for each task
sentiment_analysis = pipeline("sentiment-analysis")
text_generation = pipeline("text-generation")
translation = pipeline("translation", model="Helsinki-NLP/opus-mt-tc-big-en-ko")  # Example English to Korean translation
summarization = pipeline("summarization")
ner = pipeline("ner")

# Function to perform the chosen task
def perform_task(text, user_choice):
  if user_choice == "1":
    sentiment = sentiment_analysis(text)
    print("Sentiment Analysis:", sentiment[0]["label"])
  elif user_choice == "2":
    generated_text = text_generation(text, max_length=100, num_return_sequences=1)[0]["generated_text"]
    print("Text Generation:", generated_text)
  elif user_choice == "3":
    translated_text = translation(text, src_lang="en", tgt_lang="ko")[0]["translation"]
    print("Translation (English to Korean):", translated_text)
  elif user_choice == "4":
    summary = summarization(text, max_length=100)[0]["summary_text"]
    print("Summary:", summary)
  elif user_choice == "5":
    entities = ner(text)
    print("Named Entities:", entities)
  else:
    print("Invalid choice. Please try again.")

# Get user input text
text = input("Enter your text: ")

# Display task options and get user choice
print("\nSelect an action to perform:")
print("1. Sentiment Analysis")
print("2. Text Generation")
print("3. Translation (English to French)")
print("4. Summarization")
print("5. Named Entity Recognition (NER)")
user_choice = input("Enter your choice (1-5): ")

# Perform the chosen task
perform_task(text, user_choice)
```
